### **6. Daily Dialog**

```python
emotion_map = {
    0: 'neutral',
    1: 'anger',
    2: 'disgust',
    3: 'fear',
    4: 'joy',        
    5: 'sadness',
    6: 'surprise'
}
```

```python
import pandas as pd

BASE_PATH = '/kaggle/input/datasets/dikshachakraborty01/dailydialog-dataset'

def load_split(text_file, emotion_file, split_name):
    
    with open(f"{BASE_PATH}/{text_file}", encoding='utf-8') as f:
        texts = f.readlines()
        
    with open(f"{BASE_PATH}/{emotion_file}", encoding='utf-8') as f:
        emotions = f.readlines()
    
    data = []
    
    for t, e in zip(texts, emotions):
        
        utterances = t.strip().split('__eou__')[:-1]
        emo_labels = list(map(int, e.strip().split()))
        
        # Safety check
        if len(utterances) != len(emo_labels):
            continue
        
        for utt, emo in zip(utterances, emo_labels):
            data.append([utt.strip(), emotion_map[emo]])
    
    df = pd.DataFrame(data, columns=['text', 'emotion'])
    df['source'] = 'DailyDialog'
    df['split'] = split_name
    
    return df
```

```python
daily_train_df = load_split(
    'dialogues_train.txt',
    'dialogues_emotion_train.txt',
    'train'
)

daily_val_df = load_split(
    'dialogues_validation.txt',
    'dialogues_emotion_validation.txt',
    'validation'
)

daily_test_df = load_split(
    'dialogues_test.txt',
    'dialogues_emotion_test.txt',
    'test'
)
```

### **8. EmoryNLP**

```python
import json

emory_train_path = "/kaggle/input/datasets/dikshachakraborty01/emorynlp/emotion-detection-trn.json"
emory_dev_path   = "/kaggle/input/datasets/dikshachakraborty01/emorynlp/emotion-detection-dev.json"
emory_test_path  = "/kaggle/input/datasets/dikshachakraborty01/emorynlp/emotion-detection-tst.json"

with open(emory_train_path, "r") as f:
    emory_train_data = json.load(f)

with open(emory_dev_path, "r") as f:
    emory_dev_data = json.load(f)

with open(emory_test_path, "r") as f:
    emory_test_data = json.load(f)

print("Train samples:", len(emory_train_data))
```

```python
import pandas as pd

def flatten_emory(data, split_name):
    rows = []

    for episode in data['episodes']:
        for scene in episode['scenes']:
            for utt in scene['utterances']:
                
                text = utt['transcript'].strip()
                emotion = utt['emotion'].lower().strip()

                rows.append({
                    "text": text,
                    "emotion": emotion,
                    "source": "EmoryNLP",
                    "split": split_name
                })

    return pd.DataFrame(rows)
```

```python
emory_train_df = flatten_emory(emory_train_data, "train")
emory_dev_df   = flatten_emory(emory_dev_data, "dev")
emory_test_df  = flatten_emory(emory_test_data, "test")
```

### **3. Emobank**

Normalization

```python
df_emobank['valence'] = df_emobank['V'] / 5
df_emobank['arousal'] = df_emobank['A'] / 5
```

Mapping VAD scores to discrete emotions

```python
def map_va_to_emotion(v, a):
    if v > 0.6 and a > 0.6:
        return "joy"
    elif v < 0.4 and a > 0.6:
        return "anger"
    elif v < 0.4 and a <= 0.4:
        return "sadness"
    elif v > 0.6 and a <= 0.4:
        return "love"
    else:
        return "neutral"

df_emobank['emotion'] = df_emobank.apply(
    lambda row: map_va_to_emotion(row['valence'], row['arousal']), axis=1
)

df_emobank = df_emobank[['text','emotion']]
df_emobank['source'] = 'Emobank'
```

# **Merging of Datasets**
---

```python
all_datasets = [
    twitter_df,
    df_cbet,
    df_emobank,
    df_isear,
    df_val_ar,
    daily_df,
    meld_all,
    emory_df,
    emotionsnlp_df,
    emoint_df
]

combined_df = pd.concat(all_datasets, ignore_index=True)
```

### **Removing Duplicates from the Merged Dataset**

```python
combined_df = combined_df.drop_duplicates(subset=['text'])
combined_df = combined_df.dropna(subset=['text','emotion'])
combined_df = combined_df[combined_df['text'].str.len() > 2]
```

## **Total Emotions in the Unified Dataset**

```python
print("Total samples:", combined_df.shape[0])
print("Unique emotions:", combined_df['emotion'].nunique())
print(sorted(combined_df['emotion'].unique()))
```

### **Checking for Occurances of Emotions**

```python
emotion_presence = combined_df.groupby('emotion')['source'].nunique().sort_values(ascending=False)

print(emotion_presence)
```

## **Common vs Uncommon Emotions across Different Datasets**

```python
num_sources = combined_df['source'].nunique()

common_emotions = emotion_presence[emotion_presence == num_sources].index.tolist()
rare_emotions   = emotion_presence[emotion_presence < num_sources].index.tolist()

print("Common across ALL datasets:", common_emotions)
print("Not common:", rare_emotions)
```

### **Mapping similar emotions**

```python
harmonize_map = {
    'scared': 'fear'
}

combined_df['emotion'] = combined_df['emotion'].replace(harmonize_map)
```

### **Removing unwanted emotions**

```python
core_emotions = [
    'joy',
    'sadness',
    'anger',
    'fear',
    'disgust',
    'surprise',
    'neutral'
]

core_df = combined_df[
    combined_df['emotion'].isin(core_emotions)
].copy()
```

## **Emotion Sampling**

```python
max_per_group = 9000  

domain_balanced_df = core_df.groupby(['emotion', 'source']).apply(
    lambda x: x.sample(min(len(x), max_per_group), random_state=42)
).reset_index(drop=True)
```

```python
min_count = domain_balanced_df['emotion'].value_counts().min()

final_df = domain_balanced_df.groupby('emotion').apply(
    lambda x: x.sample(min_count, random_state=42)
).reset_index(drop=True)
```
